```python
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
df['HS'] = df.HS.str.replace("*","")
df.head()
```

```python
df.columns
```

```python
ss = StandardScaler()
df_scaled = ss.fit_transform(df)
#df_scaled.head()
```

```python
df1 = pd.DataFrame(df_scaled,columns=['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0','Exp'] )
```

```python
df1.head()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
hc = linkage(df1,method="complete")
```

```python
plt.figure(figsize=(20,12))
dendrogram(hc)
plt.show()
```

```python
from scipy.cluster.hierarchy import cut_tree
```

```python
cluster_label = cut_tree(hc,n_clusters =3)
```

```python
cl = cluster_label.reshape(-1,)
```

```python
df_hc = df.copy()
```

```python
df['Players'] = player
```

```python
df['Cluster'] = cl
```

```python
df.head()
```

```python
sns.scatterplot(x="Inns",y="Runs",hue="Cluster",data=df)
```

```python
sns.scatterplot(x="Exp",y="SR",hue="Cluster",data=df)
```

```python
sns.scatterplot(x="Runs",y="BF",hue="Cluster",data=df)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df_hc,cl)
```

```python
cluster_label = cut_tree(hc,n_clusters =2)
cl = cluster_label.reshape(-1,)
df['Cluster'] = cl
```

```python
silhouette_score(df_hc,cl)
```

```python
ss =[]
for i in range(2,11):
    cluster_label = cut_tree(hc,n_clusters =i)
    cl = cluster_label.reshape(-1,)
    df['Cluster'] = cl
    ss.append(silhouette_score(df_hc,cl))
```

```python
plt.plot(range(2,11),ss)
```

```python

hc = linkage(df1,method="single")
ss =[]
for i in range(2,11):
    cluster_label = cut_tree(hc,n_clusters =i)
    cl = cluster_label.reshape(-1,)
    df['Cluster'] = cl
    ss.append(silhouette_score(df_hc,cl))
plt.plot(range(2,11),ss)
```
